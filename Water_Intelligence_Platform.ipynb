```python
import pandas as pd
import os

# Define the path to the simulated raw data
raw_data_dir = 'data/raw'
simulated_data_path = os.path.join(raw_data_dir, 'simulated_water_data.csv')

# --- Instruction 4: Load the generated simulated_water_data.csv into a pandas DataFrame ---
df = pd.read_csv(simulated_data_path)

# Convert 'Timestamp' column to datetime objects
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# --- Instruction 5: Inspect the loaded DataFrame ---
print("First 5 rows of the DataFrame:")
print(df.head())

print("\nDataFrame Info:")
df.info()

print("\nMissing values count per column:")
print(df.isnull().sum())

print("\nDescriptive Statistics:")
print(df.describe())
```

**Reasoning**:
I need to address missing values in the DataFrame. For numerical time-series columns ('Water flow rate', 'Temperature', 'Production volume'), I will use linear interpolation within each 'Facility ID' group to maintain operational context. For the categorical 'Shift / Day / Season' column, I'll impute missing values by back-filling the previous valid observation within each facility, assuming shifts are somewhat consistent, and then filling any remaining NaNs (e.g., at the very start of a facility's data) with the mode. Before imputation, I will set 'Timestamp' as the index and sort the DataFrame to ensure proper time-series operations.



```python
print("\nShape before handling missing values:", df.shape)
print("Missing values before handling:\n", df.isnull().sum())

# Set 'Timestamp' as index and sort for time-series operations
df = df.set_index('Timestamp').sort_values(by=['Facility ID', 'Timestamp'])

# Group by 'Facility ID' and apply interpolation for numerical columns
# For 'Water flow rate', 'Temperature', 'Production volume' (numerical time-series)
df['Water flow rate'] = df.groupby('Facility ID')['Water flow rate'].transform(lambda group: group.interpolate(method='linear'))
df['Temperature'] = df.groupby('Facility ID')['Temperature'].transform(lambda group: group.interpolate(method='linear'))
df['Production volume'] = df.groupby('Facility ID')['Production volume'].transform(lambda group: group.interpolate(method='linear'))

# For 'Shift / Day / Season' (categorical)
# First, forward fill within each group
df['Shift / Day / Season'] = df.groupby('Facility ID')['Shift / Day / Season'].transform(lambda group: group.ffill())
# Then, fill any remaining NaNs (which would be at the start of a series) with the mode of that facility's shifts
for facility in df['Facility ID'].unique():
    mode_shift = df[df['Facility ID'] == facility]['Shift / Day / Season'].mode()[0]
    df.loc[(df['Facility ID'] == facility) & (df['Shift / Day / Season'].isnull()), 'Shift / Day / Season'] = mode_shift

# Ensure all interpolations and fills are complete
df = df.fillna(method='ffill') # Fallback for any remaining NaNs, though group operations should handle most
df = df.fillna(method='bfill') # Backward fill to catch any remaining leading NaNs

print("\nShape after handling missing values:", df.shape)
print("Missing values after handling:\n", df.isnull().sum())
print("\nFirst 5 rows of the DataFrame after missing value imputation:")
print(df.head())
```

Isolation Forest is suitable for high-dimensional datasets and can identify anomalies without assuming a specific data distribution. I will then categorize these outliers, focusing on 'Water flow rate' to distinguish between sensor faults (short, sharp spikes) and sustained anomalies, which could indicate real issues like leaks.



```python
from sklearn.ensemble import IsolationForest

print("\nShape before outlier detection:", df.shape)

# Outlier detection using Isolation Forest for each facility
for facility_id in df['Facility ID'].unique():
    facility_df = df[df['Facility ID'] == facility_id].copy()

    # Apply Isolation Forest to relevant numerical columns
    # 'Water flow rate', 'Temperature', 'Pressure', 'Production volume'
    numerical_cols = ['Water flow rate', 'Temperature', 'Pressure', 'Production volume']
    model = IsolationForest(random_state=42, contamination='auto') # contamination='auto' estimates proportion of outliers
    facility_df['outlier_score'] = model.fit_predict(facility_df[numerical_cols])

    # -1 indicates an outlier, 1 indicates an inlier
    facility_df['is_outlier'] = (facility_df['outlier_score'] == -1).astype(int)

    # Distinguish between sensor faults and real anomalies for 'Water flow rate'
    # A simple approach: sensor faults are short, sharp spikes; real anomalies are more sustained
    facility_df['anomaly_type'] = 'none'
    water_rate_outliers = facility_df[(facility_df['is_outlier'] == 1) & (facility_df['Water flow rate'] > facility_df['Water flow rate'].mean() + 3 * facility_df['Water flow rate'].std())]

    for i in water_rate_outliers.index:
        # Check if it's a short spike (sensor fault)
        # Look at the previous and next values. If they are normal, it's likely a sensor fault.
        if i > facility_df.index.min() and i < facility_df.index.max():
            prev_val = facility_df.loc[facility_df.index < i, 'Water flow rate'].iloc[-1] if not facility_df.loc[facility_df.index < i, 'Water flow rate'].empty else None
            next_val = facility_df.loc[facility_df.index > i, 'Water flow rate'].iloc[0] if not facility_df.loc[facility_df.index > i, 'Water flow rate'].empty else None

            # Define 'normal' range based on overall distribution or rolling average
            mean_water_rate = facility_df['Water flow rate'].mean()
            std_water_rate = facility_df['Water flow rate'].std()
            upper_bound_normal = mean_water_rate + 2 * std_water_rate

            if prev_val is not None and next_val is not None and \
               prev_val < upper_bound_normal and next_val < upper_bound_normal:
                facility_df.loc[i, 'anomaly_type'] = 'sensor_fault'
            else:
                # If not a short spike, it could be a sustained anomaly (leak, process issue)
                # This is a simplification; a more robust method would involve window analysis
                facility_df.loc[i, 'anomaly_type'] = 'sustained_anomaly'
        else:
             facility_df.loc[i, 'anomaly_type'] = 'sustained_anomaly'

    # Update the main DataFrame
    df.loc[facility_df.index, ['outlier_score', 'is_outlier', 'anomaly_type']] = facility_df[['outlier_score', 'is_outlier', 'anomaly_type']]

print("\nShape after outlier detection:", df.shape)
print("Outlier count:", df['is_outlier'].sum())
print("Anomaly types distribution:")
print(df['anomaly_type'].value_counts())
print("\nFirst 5 rows of the DataFrame after outlier detection and categorization:")
print(df.head())
```

```python
print("\n--- Visualizing Potential Cost Leakage (Instruction 7a) ---\n")

estimated_water_loss_liters = {}

for facility in facilities:
    facility_df = df[df['Facility ID'] == facility].copy()

    # Calculate a baseline water flow rate for comparison during anomalies
    # Using rolling mean of non-anomaly periods, or overall mean if no rolling data yet
    facility_df['baseline_flow'] = facility_df['Water flow rate'].rolling(window=24, min_periods=1).mean().fillna(facility_df['Water flow rate'].mean())

    # Identify sustained anomalies
    sustained_anomalies = facility_df[facility_df['anomaly_type'] == 'sustained_anomaly']

    if not sustained_anomalies.empty:
        # Calculate excess water flow rate during sustained anomalies
        # Ensure excess is not negative, meaning we only consider flow above baseline
        excess_flow = (sustained_anomalies['Water flow rate'] - sustained_anomalies['baseline_flow']).clip(lower=0)

        # Sum the excess flow for the duration of the anomaly
        # Assuming each row represents an hour (hourly data frequency)
        total_excess_water_liters = excess_flow.sum() # sum of L/h * h = Liters
        estimated_water_loss_liters[facility] = total_excess_water_liters

        print(f"Estimated water loss due to 'sustained_anomaly' for {facility}: {total_excess_water_liters:.2f} Liters")
    else:
        print(f"No 'sustained_anomaly' periods found for {facility}.")

print("\nCalculation of estimated water loss in liters for sustained anomaly periods completed for each facility.")
```

**Reasoning**:

This will involve calculating the cost of excess water and then plotting its cumulative sum.



```python
print("\n--- Plotting Cumulative Estimated Cost of Water Loss (Instruction 7b) ---\n")

for facility in facilities:
    facility_df = df[df['Facility ID'] == facility].copy()

    # Recalculate baseline flow for this facility to ensure consistency
    facility_df['baseline_flow'] = facility_df['Water flow rate'].rolling(window=24, min_periods=1).mean().fillna(facility_df['Water flow rate'].mean())

    # Identify sustained anomalies
    sustained_anomalies_df = facility_df[facility_df['anomaly_type'] == 'sustained_anomaly']

    if not sustained_anomalies_df.empty:
        # Calculate excess water flow rate during sustained anomalies
        excess_flow = (sustained_anomalies_df['Water flow rate'] - sustained_anomalies_df['baseline_flow']).clip(lower=0)

        # Calculate the cost of this excess flow (assuming Cost per unit water is per liter)
        # Note: Cost per unit water is a constant in this simulated data, but could vary.
        cost_of_excess_water = excess_flow * sustained_anomalies_df['Cost per unit water']

        # Create a DataFrame for cumulative cost, aligning it with the original time index
        cumulative_cost_series = pd.Series(0.0, index=facility_df.index) # Initialize with zeros
        cumulative_cost_series.loc[cost_of_excess_water.index] = cost_of_excess_water # Assign costs at anomaly points

        # Calculate cumulative sum, filling NaNs (where no anomaly occurred) with previous value
        facility_df['cumulative_cost_of_loss'] = cumulative_cost_series.cumsum()

        plt.figure(figsize=(15, 6))
        plt.plot(facility_df.index, facility_df['cumulative_cost_of_loss'], label=f'Cumulative Cost of Water Loss for {facility}', color='red', alpha=0.8)
        plt.title(f'Cumulative Estimated Cost of Water Loss Over Time for {facility}')
        plt.xlabel('Timestamp')
        plt.ylabel('Cumulative Cost ($)')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    else:
        print(f"No 'sustained_anomaly' periods with calculated cost for {facility}.")

print("Cumulative estimated cost of water loss plots generated for each facility.")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

print("\n--- Building Baseline Water Demand Models (Instruction 5) ---\n")

# Get unique facilities from the 'normal' DataFrame
facilities = df_normal['Facility ID'].unique()

# Dictionary to store models and their predictions
baseline_models = {}
predicted_water_flow = pd.Series(index=df.index, dtype=float)

for facility in facilities:
    print(f"\nBuilding model for {facility}...")
    facility_df_normal = df_normal[df_normal['Facility ID'] == facility].copy()

    # Prepare features (X) and target (y)
    # Time-based features: hour_of_day, day_of_week, month, is_peak_hours
    # Operational features: Production volume, Cleaning cycle flag

    # Add hour_of_day if not already present
    if 'hour_of_day' not in facility_df_normal.columns:
        facility_df_normal['hour_of_day'] = facility_df_normal.index.hour

    # Select features that are available in df_normal
    features = [
        'hour_of_day',
        'day_of_week',
        'month',
        'is_peak_hours',
        'Production volume',
        'Cleaning cycle flag'
    ]

    # Filter features to only include those present in the DataFrame
    available_features = [f for f in features if f in facility_df_normal.columns]

    # Convert categorical features if necessary (though current features are already numerical or binary)
    # Example for 'Shift / Day / Season' if it were used directly:
    # X = pd.get_dummies(facility_df_normal[available_features + ['Shift / Day / Season']], columns=['Shift / Day / Season'], drop_first=True)

    X = facility_df_normal[available_features]
    y = facility_df_normal['Water flow rate']

    # Ensure no NaN values in selected features or target after filtering 'normal' data
    # Although previous steps should have handled this, it's good practice for model training
    X = X.dropna()
    y = y.loc[X.index] # Align y with X after dropping NaNs

    if X.empty:
        print(f"No sufficient data for training model for {facility} after dropping NaNs.")
        continue

    # Train a Linear Regression model
    model = LinearRegression()
    model.fit(X, y)

    # Store the model
    baseline_models[facility] = model

    # Predict on the entire original (preprocessed) facility data to get expected flow for all points
    # Create features for the *full* facility data (df, not df_normal)
    full_facility_df = df[df['Facility ID'] == facility].copy()
    if 'hour_of_day' not in full_facility_df.columns:
        full_facility_df['hour_of_day'] = full_facility_df.index.hour

    X_full = full_facility_df[available_features].dropna() # Drop NaNs in features for prediction

    # Only predict where features are available
    predicted_values = pd.Series(model.predict(X_full), index=X_full.index)
    predicted_water_flow.loc[predicted_values.index] = predicted_values

    # Evaluate the model on the 'normal' training data
    y_pred_normal = model.predict(X)
    r2 = r2_score(y, y_pred_normal)
    print(f"R-squared for {facility} on normal data: {r2:.4f}")

# Add the predicted water flow to the original DataFrame
df['predicted_water_flow'] = predicted_water_flow

# Fill any remaining NaNs (e.g., where features for prediction were missing) in predicted_water_flow
# with the mean water flow rate for that facility, or 0 if appropriate
df['predicted_water_flow'] = df.groupby('Facility ID')['predicted_water_flow'].transform(lambda x: x.fillna(x.mean()))
df['predicted_water_flow'] = df['predicted_water_flow'].fillna(0) # Final fallback

print("\nBaseline models built and predictions added to DataFrame.")
print("Updated DataFrame head with 'predicted_water_flow' column:")
print(df.head())
print("DataFrame info after adding predicted water flow:")
df.info()
```

**Reasoning**:
Now that time-series plots of residuals have been generated, I will continue with instruction 8 by performing statistical tests for normality (Shapiro-Wilk) and generating Q-Q plots for each facility's residuals to visually inspect their distribution against a normal distribution.



```python
from scipy.stats import shapiro
import statsmodels.api as sm

print("\n--- Checking for Normality of Residuals (Instruction 8) ---\n")

for facility in facilities:
    facility_df = df[df['Facility ID'] == facility].copy()
    residuals = facility_df['residuals'].dropna() # Drop NaNs if any, for test

    if len(residuals) > 5000: # Shapiro-Wilk test is only valid for small sample sizes, using it for illustrative purposes. For larger datasets, consider alternative normality tests or focus on Q-Q plots.
        print(f"\nSkipping Shapiro-Wilk test for {facility} due to large sample size. (len={len(residuals)} > 5000)")
    else:
        # Shapiro-Wilk test for normality
        stat, p = shapiro(residuals)
        print(f'\nShapiro-Wilk Test for {facility}:')
        print(f'Statistics=%.3f, p=%.3f' % (stat, p))
        alpha = 0.05
        if p > alpha:
            print('Sample looks Gaussian (fail to reject H0)')
        else:
            print('Sample does not look Gaussian (reject H0)')

    # Q-Q plot for normality visual inspection
    plt.figure(figsize=(8, 6))
    sm.qqplot(residuals, line='s', ax=plt.gca())
    plt.title(f'Q-Q Plot of Residuals for {facility}')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

print("Shapiro-Wilk tests and Q-Q plots for residuals generated for each facility.")
```

**Reasoning**:
Now that homoscedasticity has been visually inspected, I will complete the model assumption validation by checking for autocorrelation in the residuals using the Ljung-Box test for each facility, as specified in instruction 8.



```python
from statsmodels.stats.diagnostic import acorr_ljungbox

print("\n--- Checking for Autocorrelation of Residuals (Instruction 8) ---\n")

for facility in facilities:
    facility_df = df[df['Facility ID'] == facility].copy()
    residuals = facility_df['residuals'].dropna() # Drop NaNs if any, for test

    if residuals.empty:
        print(f"No residuals available for Ljung-Box test for {facility}.")
        continue

    # Ljung-Box test for autocorrelation
    # lags: number of lags to test for autocorrelation. Hourly data, check up to 24 hours.
    # boxpierce: return only Ljung-Box statistic and p-value
    lb_test = acorr_ljungbox(residuals, lags=[24], return_df=True)

    print(f'\nLjung-Box Test for Autocorrelation for {facility} (up to 24 lags):')
    print(lb_test)

    # Interpret results for the first lag in the list (24 in this case)
    p_value_24 = lb_test['lb_pvalue'].iloc[0]
    alpha = 0.05
    if p_value_24 < alpha:
        print(f'Autocorrelation present at lag 24 (p={p_value_24:.4f} < {alpha}). Residuals are not white noise.')
    else:
        print(f'No significant autocorrelation at lag 24 (p={p_value_24:.4f} >= {alpha}). Residuals are close to white noise.')

print("Ljung-Box tests for autocorrelation of residuals generated for each facility.")
```

**Reasoning**:
Now that the data is loaded and includes `predicted_water_flow` and `residuals`, I will implement residual-based anomaly detection as specified in the subtask. This involves defining a threshold for residuals to flag significant deviations from the baseline, and then categorizing these anomalies.



```python
import matplotlib.pyplot as plt
import seaborn as sns

print("\n--- Implementing Residual-Based Anomaly Detection (Instruction 4) ---\n")

# Define anomaly thresholds based on residuals
# For simplicity, we'll use a multiple of the standard deviation of residuals
# for each facility.
# A more sophisticated approach would involve dynamic thresholds or other statistical methods.

# Initialize columns for new anomaly detection results
df['new_anomaly_flag'] = 0
df['new_anomaly_type'] = 'normal'
df['anomaly_severity'] = 'none'

for facility in facilities:
    facility_df = df[df['Facility ID'] == facility].copy()

    # Calculate standard deviation of residuals for this facility
    # We'll use residuals from 'normal' periods or the overall residuals for simplicity here.
    # Using overall residuals to catch all potential anomalies.
    residuals_std = facility_df['residuals'].std()

    # Define thresholds (e.g., 2 or 3 standard deviations)
    # Positive residuals above threshold might indicate unexpected high usage (leakage, over-cleaning)
    # Negative residuals below threshold might indicate unexpected low usage (sensor error, process halt)
    upper_threshold = 2.5 * residuals_std
    lower_threshold = -2.5 * residuals_std

    # Flag anomalies based on residuals
    facility_df.loc[facility_df['residuals'] > upper_threshold, 'new_anomaly_flag'] = 1
    facility_df.loc[facility_df['residuals'] < lower_threshold, 'new_anomaly_flag'] = 1

    # Categorize anomalies
    # If water flow rate is significantly higher than predicted
    # and not explained by cleaning cycle, could be leakage or inefficiency.
    # If water flow rate is significantly lower than predicted, could be sensor error or unexpected shutdown.

    # High residuals (actual > predicted)
    high_residual_anomalies = facility_df[
        (facility_df['new_anomaly_flag'] == 1) &
        (facility_df['residuals'] > upper_threshold)
    ]

    # Low residuals (actual < predicted)
    low_residual_anomalies = facility_df[
        (facility_df['new_anomaly_flag'] == 1) &
        (facility_df['residuals'] < lower_threshold)
    ]

    # Categorize high residuals
    for idx in high_residual_anomalies.index:
        if facility_df.loc[idx, 'Cleaning cycle flag'] == 1: # High flow during cleaning
            # If it's a very high flow during cleaning, might be 'over-cleaning' or just normal cleaning variation
            if facility_df.loc[idx, 'residuals'] > 3.5 * residuals_std: # More severe than normal high
                 facility_df.loc[idx, 'new_anomaly_type'] = 'over_cleaning_event'
            else:
                facility_df.loc[idx, 'new_anomaly_type'] = 'normal_cleaning_variation'
        else: # High flow not during cleaning
            # Could be a leak or operational inefficiency
            facility_df.loc[idx, 'new_anomaly_type'] = 'potential_leak_or_inefficiency'

        # Add severity tagging based on residual magnitude (e.g., 2.5-3.5 std = moderate, >3.5 std = severe)
        if facility_df.loc[idx, 'residuals'] > 3.5 * residuals_std:
            facility_df.loc[idx, 'anomaly_severity'] = 'severe'
        elif facility_df.loc[idx, 'residuals'] > 2.5 * residuals_std:
            facility_df.loc[idx, 'anomaly_severity'] = 'moderate'

    # Categorize low residuals
    for idx in low_residual_anomalies.index:
        # These are cases where actual usage is much lower than predicted
        facility_df.loc[idx, 'new_anomaly_type'] = 'potential_sensor_error_or_shutdown'

        # Severity for low residuals (absolute value)
        if abs(facility_df.loc[idx, 'residuals']) > 3.5 * residuals_std:
            facility_df.loc[idx, 'anomaly_severity'] = 'severe'
        elif abs(facility_df.loc[idx, 'residuals']) > 2.5 * residuals_std:
            facility_df.loc[idx, 'anomaly_severity'] = 'moderate'

    # Update the main DataFrame with the new anomaly flags and types
    df.loc[facility_df.index, ['new_anomaly_flag', 'new_anomaly_type', 'anomaly_severity']] = facility_df[['new_anomaly_flag', 'new_anomaly_type', 'anomaly_severity']]


print("Anomaly detection and categorization completed.")
print("\nFirst 5 rows of DataFrame with new anomaly columns:")
print(df[['Water flow rate', 'predicted_water_flow', 'residuals', 'new_anomaly_flag', 'new_anomaly_type', 'anomaly_severity']].head())

print("\nDistribution of new anomaly types:")
print(df['new_anomaly_type'].value_counts())

print("\nDistribution of anomaly severity:")
print(df['anomaly_severity'].value_counts())
```

```python
import pandas as pd
import os

# Define the path to the preprocessed data
processed_data_dir = 'data/processed'
preprocessed_data_path = os.path.join(processed_data_dir, 'preprocessed_water_data.csv')

# --- Instruction 3: Load the preprocessed_water_data.csv file into a pandas DataFrame. ---
df = pd.read_csv(preprocessed_data_path)

# --- Instruction 4: Convert the 'Timestamp' column to datetime objects and set it as the DataFrame's index, then sort the index. ---
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df = df.set_index('Timestamp').sort_index()

print(f"Loaded preprocessed data from: {preprocessed_data_path}")

# --- Instruction 5: Display the first 5 rows of the DataFrame, its information, and descriptive statistics. ---
print("\nFirst 5 rows of the DataFrame:")
print(df.head())

print("\nDataFrame Info:")
df.info()

print("\nDescriptive Statistics:")
print(df.describe())
```

**Reasoning**:
The current DataFrame `df` does not contain the `predicted_water_flow` and `residuals` columns which are crucial for impact quantification. These columns were generated in the previous `03_baseline_demand_model.ipynb` notebook. Therefore, I need to recalculate them by reloading the preprocessed data, re-establishing the 'normal' water behavior, and re-training the baseline model, then calculate residuals. This ensures the current notebook has all necessary data for the subtask.



```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Recalculate df_normal by filtering out periods identified as 'sustained_anomaly' or 'sensor_fault'
# This is done to ensure a clean baseline for the model, as these were anomalies identified in the EDA phase.
df_normal = df[~df['anomaly_type'].isin(['sustained_anomaly', 'sensor_fault', 'normal_cleaning_variation', 'over_cleaning_event'])].copy()

# Get unique facilities from the 'normal' DataFrame
facilities = df_normal['Facility ID'].unique()

# Dictionary to store models and their predictions
baseline_models = {}
predicted_water_flow_series = pd.Series(index=df.index, dtype=float)

for facility in facilities:
    print(f"Recalculating baseline model predictions for {facility}...")
    facility_df_normal = df_normal[df_normal['Facility ID'] == facility].copy()

    # Add hour_of_day if not already present, as it's a feature used in the model
    if 'hour_of_day' not in facility_df_normal.columns:
        facility_df_normal['hour_of_day'] = facility_df_normal.index.hour

    # Select features that were used for baseline model training
    features = [
        'hour_of_day',
        'day_of_week',
        'month',
        'is_peak_hours',
        'Production volume',
        'Cleaning cycle flag'
    ]

    # Filter features to only include those present in the DataFrame (some might have been dropped or renamed)
    available_features = [f for f in features if f in facility_df_normal.columns]

    X = facility_df_normal[available_features]
    y = facility_df_normal['Water flow rate']

    # Drop any remaining NaN values in selected features or target before training
    # This is important for model fitting stability.
    X = X.dropna()
    y = y.loc[X.index] # Align y with X after dropping NaNs

    if X.empty:
        print(f"No sufficient data for training model for {facility} after dropping NaNs.")
        continue

    # Train a Linear Regression model
    model = LinearRegression()
    model.fit(X, y)

    # Store the trained model
    baseline_models[facility] = model

    # Predict on the entire original (preprocessed) facility data (df, not df_normal) to get expected flow for all points
    full_facility_df = df[df['Facility ID'] == facility].copy()
    if 'hour_of_day' not in full_facility_df.columns:
        full_facility_df['hour_of_day'] = full_facility_df.index.hour

    X_full = full_facility_df[available_features].dropna() # Drop NaNs in features for prediction

    # Only predict where features are available
    predicted_values = pd.Series(model.predict(X_full), index=X_full.index)
    predicted_water_flow_series.loc[predicted_values.index] = predicted_values

# Add the predicted water flow to the original DataFrame
df['predicted_water_flow'] = predicted_water_flow_series

# Fill any remaining NaNs in 'predicted_water_flow' (e.g., where features for prediction were missing)
# with the mean predicted water flow rate for that facility, or 0 as a final fallback.
df['predicted_water_flow'] = df.groupby('Facility ID')['predicted_water_flow'].transform(lambda x: x.fillna(x.mean()))
df['predicted_water_flow'] = df['predicted_water_flow'].fillna(0)

# Calculate residuals: actual water flow rate - predicted water flow rate
df['residuals'] = df['Water flow rate'] - df['predicted_water_flow']

print("Predicted water flow and residuals columns added to DataFrame.")

print("\nFirst 5 rows of DataFrame with new predicted_water_flow and residuals columns:")
print(df[['Water flow rate', 'predicted_water_flow', 'residuals']].head())

print("\nDataFrame Info after adding predicted water flow and residuals:")
df.info()
```

**Reasoning**:
Now that the data is loaded and includes `predicted_water_flow` and `residuals`, I will implement residual-based anomaly detection as specified in the subtask. This involves defining a threshold for residuals to flag significant deviations from the baseline, and then categorizing these anomalies.



```python
import matplotlib.pyplot as plt
import seaborn as sns

print("\n--- Implementing Residual-Based Anomaly Detection (Instruction 4) ---\n")

# Define anomaly thresholds based on residuals
# For simplicity, we'll use a multiple of the standard deviation of residuals
# for each facility.
# A more sophisticated approach would involve dynamic thresholds or other statistical methods.

# Initialize columns for new anomaly detection results
df['new_anomaly_flag'] = 0
df['new_anomaly_type'] = 'normal'
df['anomaly_severity'] = 'none'

for facility in facilities:
    facility_df = df[df['Facility ID'] == facility].copy()

    # Calculate standard deviation of residuals for this facility
    # We'll use residuals from 'normal' periods or the overall residuals for simplicity here.
    # Using overall residuals to catch all potential anomalies.
    residuals_std = facility_df['residuals'].std()

    # Define thresholds (e.g., 2 or 3 standard deviations)
    # Positive residuals above threshold might indicate unexpected high usage (leakage, over-cleaning)
    # Negative residuals below threshold might indicate unexpected low usage (sensor error, process halt)
    upper_threshold = 2.5 * residuals_std
    lower_threshold = -2.5 * residuals_std

    # Flag anomalies based on residuals
    facility_df.loc[facility_df['residuals'] > upper_threshold, 'new_anomaly_flag'] = 1
    facility_df.loc[facility_df['residuals'] < lower_threshold, 'new_anomaly_flag'] = 1

    # Categorize anomalies
    # If water flow rate is significantly higher than predicted
    # and not explained by cleaning cycle, could be leakage or inefficiency.
    # If water flow rate is significantly lower than predicted, could be sensor error or unexpected shutdown.

    # High residuals (actual > predicted)
    high_residual_anomalies = facility_df[
        (facility_df['new_anomaly_flag'] == 1) &
        (facility_df['residuals'] > upper_threshold)
    ]

    # Low residuals (actual < predicted)
    low_residual_anomalies = facility_df[
        (facility_df['new_anomaly_flag'] == 1) &
        (facility_df['residuals'] < lower_threshold)
    ]

    # Categorize high residuals
    for idx in high_residual_anomalies.index:
        if facility_df.loc[idx, 'Cleaning cycle flag'] == 1: # High flow during cleaning
            # If it's a very high flow during cleaning, might be 'over-cleaning' or just normal cleaning variation
            if facility_df.loc[idx, 'residuals'] > 3.5 * residuals_std: # More severe than normal high
                 facility_df.loc[idx, 'new_anomaly_type'] = 'over_cleaning_event'
            else:
                facility_df.loc[idx, 'new_anomaly_type'] = 'normal_cleaning_variation'
        else: # High flow not during cleaning
            # Could be a leak or operational inefficiency
            facility_df.loc[idx, 'new_anomaly_type'] = 'potential_leak_or_inefficiency'

        # Add severity tagging based on residual magnitude (e.g., 2.5-3.5 std = moderate, >3.5 std = severe)
        if facility_df.loc[idx, 'residuals'] > 3.5 * residuals_std:
            facility_df.loc[idx, 'anomaly_severity'] = 'severe'
        elif facility_df.loc[idx, 'residuals'] > 2.5 * residuals_std:
            facility_df.loc[idx, 'anomaly_severity'] = 'moderate'

    # Categorize low residuals
    for idx in low_residual_anomalies.index:
        # These are cases where actual usage is much lower than predicted
        facility_df.loc[idx, 'new_anomaly_type'] = 'potential_sensor_error_or_shutdown'

        # Severity for low residuals (absolute value)
        if abs(facility_df.loc[idx, 'residuals']) > 3.5 * residuals_std:
            facility_df.loc[idx, 'anomaly_severity'] = 'severe'
        elif abs(facility_df.loc[idx, 'residuals']) > 2.5 * residuals_std:
            facility_df.loc[idx, 'anomaly_severity'] = 'moderate'

    # Update the main DataFrame with the new anomaly flags and types
    df.loc[facility_df.index, ['new_anomaly_flag', 'new_anomaly_type', 'anomaly_severity']] = facility_df[['new_anomaly_flag', 'new_anomaly_type', 'anomaly_severity']]


print("Anomaly detection and categorization completed.")
print("\nFirst 5 rows of DataFrame with new anomaly columns:")
print(df[['Water flow rate', 'predicted_water_flow', 'residuals', 'new_anomaly_flag', 'new_anomaly_type', 'anomaly_severity']].head())

print("\nDistribution of new anomaly types:")
print(df['new_anomaly_type'].value_counts())

print("\nDistribution of anomaly severity:")
print(df['anomaly_severity'].value_counts())
```

**Reasoning**:
Now that anomalies have been detected and categorized, I will quantify the water wasted (in liters) for each anomaly, focusing on those where actual water usage exceeded the predicted baseline (e.g., potential leaks, over-cleaning). For negative residuals, which might indicate sensor errors or shutdowns, the 'wasted' liters are considered 0 for the purpose of quantifying excess water loss.



```python
print("\n--- Quantifying Liters Wasted (Instruction 5) ---\n")

# Initialize a column for liters wasted
df['liters_wasted'] = 0.0

# Iterate through each facility to calculate wasted liters for anomalies
for facility in facilities:
    facility_df = df[df['Facility ID'] == facility].copy()

    # Identify anomalies where actual water flow is higher than predicted (positive residuals)
    # These are 'potential_leak_or_inefficiency', 'over_cleaning_event', 'normal_cleaning_variation'
    # The 'residuals' directly represent the excess flow (L/h). Since data is hourly, residual is already in Liters for that hour.
    excess_water_anomalies = facility_df[
        (facility_df['new_anomaly_flag'] == 1) &
        (facility_df['residuals'] > 0)
    ]

    # Sum the positive residuals to get total wasted liters for these anomaly types
    # Each residual value is already in Liters/hour * 1 hour = Liters
    df.loc[excess_water_anomalies.index, 'liters_wasted'] = excess_water_anomalies['residuals']

    total_wasted_liters_facility = df.loc[df['Facility ID'] == facility, 'liters_wasted'].sum()
    print(f"Total estimated liters wasted for {facility}: {total_wasted_liters_facility:.2f} Liters")

print("\nQuantification of liters wasted completed.")
print("\nFirst 5 rows of DataFrame with 'liters_wasted' column:")
print(df[['Water flow rate', 'predicted_water_flow', 'residuals', 'new_anomaly_type', 'liters_wasted']].head())
```

**Reasoning**:
Now that liters wasted have been quantified, I will calculate the monetary cost impact by multiplying the 'liters_wasted' by 'Cost per unit water'. I will also define a proxy value for CO2 equivalent per liter of water and calculate the CO2 impact, then display the updated DataFrame and print totals per facility.



```python
print("\n--- Quantifying Monetary Cost and CO2 Equivalent Impact (Instruction 6) ---\n")

# --- Monetary Cost Impact (₹) ---
# Assuming 'Cost per unit water' is in $/Liter. We need to convert this to ₹/Liter.
# Let's assume an exchange rate for demonstration. For this context, let's treat it as the final currency.
# If `Cost per unit water` is already per liter, we can directly multiply.
df['monetary_cost_impact'] = df['liters_wasted'] * df['Cost per unit water']

# --- CO2 Equivalent Impact (proxy ESG metric) ---
# Define a proxy conversion factor: kg CO2e per liter of water
# This value can vary widely based on water source, treatment, and distribution.
# A common range is 0.2 to 2 grams CO2e per liter of tap water.
# Let's use 0.5 grams (0.0005 kg) CO2e per liter as a reasonable proxy for industrial usage.
co2_per_liter = 0.0005 # kg CO2e per Liter
df['co2_equivalent_impact'] = df['liters_wasted'] * co2_per_liter

for facility in facilities:
    total_wasted_liters = df.loc[df['Facility ID'] == facility, 'liters_wasted'].sum()
    total_monetary_cost = df.loc[df['Facility ID'] == facility, 'monetary_cost_impact'].sum()
    total_co2_impact = df.loc[df['Facility ID'] == facility, 'co2_equivalent_impact'].sum()

    print(f"\nSummary for {facility}:")
    print(f"  Total Liters Wasted: {total_wasted_liters:.2f} Liters")
    print(f"  Total Monetary Cost Impact: ₹{total_monetary_cost:.2f}")
    print(f"  Total CO2 Equivalent Impact: {total_co2_impact:.2f} kg CO2e")

print("\nQuantification of monetary cost and CO2 equivalent impact completed.")
print("\nFirst 5 rows of DataFrame with new impact columns:")
print(df[['Water flow rate', 'predicted_water_flow', 'residuals', 'new_anomaly_type', 'liters_wasted', 'monetary_cost_impact', 'co2_equivalent_impact']].head())
```

**Reasoning**:
Now that the monetary cost and CO2 equivalent impact have been calculated per anomaly, I will proceed with instruction 7 to estimate the monthly and annualized impact for each facility.



```python
print("\n--- Estimating Monthly and Annualized Impact (Instruction 7) ---\n")

# Create columns for year and month to facilitate aggregation
df['year'] = df.index.year
df['month'] = df.index.month

# Aggregate impact metrics by Facility ID, year, and month
monthly_impact = df.groupby(['Facility ID', 'year', 'month']).agg(
    total_liters_wasted=('liters_wasted', 'sum'),
    total_monetary_cost=('monetary_cost_impact', 'sum'),
    total_co2_impact=('co2_equivalent_impact', 'sum')
).reset_index()

print("Monthly Impact Summary:")
print(monthly_impact.head())

# Calculate annualized impact for each facility
annualized_impact = monthly_impact.groupby('Facility ID').agg(
    avg_monthly_liters_wasted=('total_liters_wasted', 'mean'),
    total_annual_liters_wasted=('total_liters_wasted', 'sum'),
    avg_monthly_monetary_cost=('total_monetary_cost', 'mean'),
    total_annual_monetary_cost=('total_monetary_cost', 'sum'),
    avg_monthly_co2_impact=('total_co2_impact', 'mean'),
    total_annual_co2_impact=('total_co2_impact', 'sum')
).reset_index()

print("\nAnnualized Impact Summary:")
print(annualized_impact)

print("\nEstimation of monthly and annualized impact completed.")
```

```python
import pandas as pd
import os

# Ensure the necessary columns for KPI calculation are present
# Reload the data if df is not the latest version with all anomaly info
# Assuming df is already loaded and contains 'Facility ID', 'Water flow rate', 'Production volume',
# 'monetary_cost_impact', 'co2_equivalent_impact', 'new_anomaly_type', 'anomaly_severity'

# Calculate KPIs for each facility
kpi_data = []

for facility_id in df['Facility ID'].unique():
    facility_df = df[df['Facility ID'] == facility_id].copy()

    # 1. Water Use Efficiency (L/Unit Production)
    # Avoid division by zero
    total_water_consumed = facility_df['Water flow rate'].sum()
    total_production_volume = facility_df['Production volume'].sum()
    water_use_efficiency = total_water_consumed / total_production_volume if total_production_volume > 0 else 0

    # 2. Cost of Avoidable Water Loss (₹) - based on positive residuals, non-sensor_error anomalies
    avoidable_anomaly_types = ['potential_leak_or_inefficiency', 'over_cleaning_event']
    cost_of_avoidable_water_loss = facility_df[
        facility_df['new_anomaly_type'].isin(avoidable_anomaly_types)
    ]['monetary_cost_impact'].sum()

    # 3. Monetary Impact of Sustained Anomalies (₹/Month) - total from all non-sensor_error anomalies
    # Assuming 'monetary_cost_impact' is already annualized or per period, need to normalize to monthly.
    # For this simulation, we'll assume the data covers a year, so we can divide by 12 for average monthly.
    total_monetary_impact_all_anomalies = facility_df[
        ~facility_df['new_anomaly_type'].isin(['potential_sensor_error_or_shutdown', 'normal'])
    ]['monetary_cost_impact'].sum()
    avg_monthly_monetary_impact = total_monetary_impact_all_anomalies / 12 # Assuming 12 months data

    # 4. Frequency of Critical Water Anomalies (Count/Month) - severe 'potential_leak_or_inefficiency'
    critical_anomalies_count = facility_df[
        (facility_df['new_anomaly_type'] == 'potential_leak_or_inefficiency') &
        (facility_df['anomaly_severity'] == 'severe')
    ].shape[0]
    avg_monthly_critical_anomalies_freq = critical_anomalies_count / 12 # Assuming 12 months data

    # 5. Total Water Consumption (L/Year)
    total_annual_water_consumption = facility_df['Water flow rate'].sum()

    # 6. Carbon Footprint from Water Usage (kg CO2e/Year)
    total_annual_co2_impact = facility_df['co2_equivalent_impact'].sum()

    kpi_data.append({
        'Facility ID': facility_id,
        'Water Use Efficiency (L/Unit)': water_use_efficiency,
        'Cost of Avoidable Water Loss (₹/Year)': cost_of_avoidable_water_loss,
        'Avg Monthly Monetary Impact (₹)': avg_monthly_monetary_impact,
        'Avg Monthly Critical Anomaly Frequency (Count)': avg_monthly_critical_anomalies_freq,
        'Total Annual Water Consumption (L)': total_annual_water_consumption,
        'Total Annual Carbon Footprint (kg CO2e)': total_annual_co2_impact
    })

kpi_tables = pd.DataFrame(kpi_data)

# Define the path to save the aggregated KPI tables
processed_data_dir = 'data/processed'
kpi_tables_path = os.path.join(processed_data_dir, 'kpi_tables.csv')

# Save the aggregated tables to CSV
kpi_tables.to_csv(kpi_tables_path, index=False)

print(f"Aggregated KPI tables saved to {kpi_tables_path}")
print("\nAggregated KPI Tables:")
print(kpi_tables.head())
```
